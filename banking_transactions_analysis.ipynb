```python

merged_data['IsHighValue'] = merged_data['Amount'] > 9000

flagged = merged_data[merged_data['IsHighValue']]
print("High-value Transactions:")
print(flagged[['TransactionID', 'FullName', 'Amount', 'Date', 'Location']].head(100))
```

```python

merged_data['Date'] = pd.to_datetime(merged_data['Date'], errors='coerce')

withdrawals = merged_data[merged_data['Type'] == 'Withdrawal'].copy()
withdrawals['Day'] = withdrawals['Date'].dt.date

suspicious_counts = withdrawals.groupby(['AccountID', 'Day']).size()

suspicious_multi = suspicious_counts[suspicious_counts >= 3].reset_index()
suspicious_multi.columns = ['AccountID', 'Day', 'SuspiciousWithdrawalCount']


merged_data['Day'] = merged_data['Date'].dt.date
merged = merged_data.merge(
    suspicious_multi, on=['AccountID', 'Day'], how='left', suffixes=('', '_sus')
)

merged['Flag_Suspicious'] = merged['SuspiciousWithdrawalCount'].fillna(0) >= 3


print("ðŸš¨ Suspicious transactions flagged:")
print(merged['Flag_Suspicious'].sum())
merged[merged['Flag_Suspicious'] == True].head(10)
```
